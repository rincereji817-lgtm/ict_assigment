```python
def celsius_to_fahrenheit(celsius):
    """Converts temperature from Celsius to Fahrenheit."""
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

# Example usage:
celsius_temp = float(input("Enter temperature in Celsius: "))
fahrenheit_temp = celsius_to_fahrenheit(celsius_temp)
print(f"{celsius_temp}°C is equal to {fahrenheit_temp}°F")
```

```python
import math

# Get the lengths of the three sides from the user
a = float(input("Enter the length of side a: "))
b = float(input("Enter the length of side b: "))
c = float(input("Enter the length of side c: "))

# Check if the given side lengths can form a valid triangle
if (a + b > c) and (a + c > b) and (b + c > a):
    # Calculate the semi-perimeter
    s = (a + b + c) / 2

    # Calculate the area using Heron's formula
    area = math.sqrt(s * (s - a) * (s - b) * (s - c))

    print(f"The area of the triangle with sides {a}, {b}, and {c} is: {area:.2f} square units")
else:
    print("Error: The given side lengths do not form a valid triangle.")
```
